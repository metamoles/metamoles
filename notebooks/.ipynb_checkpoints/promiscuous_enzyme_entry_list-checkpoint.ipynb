```python
from Bio.KEGG import REST
```

```python
enzyme = REST.kegg_list("enzyme").read()
```

```python
enzyme_entry = []
for line in enzyme.rstrip().split("\n"):
    entry, description = line.split("\t")
    enzyme_entry.append(entry)
```
