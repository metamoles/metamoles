```python
import pandas as pd
import numpy as np
```

```python
# I hack a .dat file by opening in excel and save it as .csv file which I'm more confident in
csv = pd.read_csv('enzrxns.csv', skiprows = 59, names=['Col'])
#csv.columns('Col')
```

```python
k1 = 0
ID = [] # 'UNIQUE-ID - '
ID_L = len('UNIQUE-ID - ')

k2 = 0
CName = [] # COMMON-NAME - 
CName_L = len('COMMON-NAME - ')

k3 = 0
enz = [] # ENZYME - 
enz_L = len('ENZYME - ')

k4 = 0
rxn = [] # REACTION - 
rxn_L = len('REACTION - ')
reaction = []

k5 = 0
alt = [] # ALTERNATIVE-SUBSTRATES - 
alt_L = len('ALTERNATIVE-SUBSTRATES - ')
substrate = []

k6 = 0
sub = [] # ^SUBSTRATE - 
sub_L = len('^SUBSTRATE - ')
subs = []

k7 = 0
KM = [] # KM - 
KM_L = len('KM - ')
km = []

k8 = 0
KCAT = [] # KCAT - 
KCAT_L = len('KCAT - ')
kcat = []

k9 = 0
VMAX = [] # VMAX - 
VMAX_L = len('VMAX - ')
vmax = []

k10 = 0
PH = [] # PH-OPT -
PH_L = len('PH-OPT - ')
ph = []

k11 = 0
act = [] # SPECIFIC-ACTIVITY - 
act_L = len('SPECIFIC-ACTIVITY - ')

k12 = 0
temp = [] # TEMPERATURE-OPT - 
temp_L = len('TEMPERATURE-OPT - ')





for i in range(csv.shape[0]):
    
    data = csv['Col'][i]
    if data[:ID_L] == 'UNIQUE-ID - ':
        ID.append(data[(ID_L):])
        k1 += 1
        #print(data[(ID_L):])
    elif data[:CName_L] == 'COMMON-NAME - ':
        CName.append(data[(CName_L):])
        k2 += 1
    elif data[:enz_L] == 'ENZYME - ':
        enz.append(data[(enz_L):])
        k3 += 1
    elif data[:rxn_L] == 'REACTION - ':
        reaction.append(data[(rxn_L):])
        # k4 += 1
    elif data[:alt_L] == 'ALTERNATIVE-SUBSTRATES - ':
        substrate.append(data[(alt_L):]) # there are several substrate which will be appended to another list first
        # k5 += 1
    elif data[:sub_L] == '^SUBSTRATE - ':
        subs.append(data[sub_L:]) # there might be more than one substrate
        # k6 += 1
    elif data[:KM_L] == 'KM - ':
        km.append(data[KM_L:])
        # k7 += 1
    elif data[:KCAT_L] == 'KCAT - ':
        kcat.append(data[KCAT_L:])
        # k8 += 1
    elif data[:VMAX_L] == 'VMAX - ':
        vmax.append(data[VMAX_L:])
        # k9 += 1
    elif data[:PH_L] == 'PH-OPT - ':
        ph.append(data[PH_L:])
        # k10 += 1
    elif data[:act_L] == 'SPECIFIC-ACTIVITY - ':
        act.append(data[act_L:])
        k11 += 1
    elif data[:temp_L] == 'TEMPERATURE-OPT - ':
        temp.append(data[temp_L:])
        k12 += 1
    else:
        pass
    
    if k2 < k1-1:
        CName.append('N/A')
        k2 += 1
    if k3 < k1-1:
        enz.append('N/A')
        k3 += 1
    if k4 < k1-1:
        if reaction != []:
            rxn.append(reaction)
            reaction = []
        else:
            rxn.append('N/A')
        k4 += 1
    if k5 < k1-1:
        #print(substrate)
        if substrate != []:
            alt.append(substrate)
            substrate = []
        else:
            alt.append('N/A')
        k5 += 1
    if k6 < k1-1:
        if subs != []:
            sub.append(subs)
            subs = []
        else:
            sub.append('N/A')
        k6 += 1
    if k7 < k1-1:
        if km != []:
            KM.append(km)
            km = []
        else:
            KM.append('N/A')
        k7 += 1
    if k8 < k1-1:
        if kcat != []:
            KCAT.append(kcat)
            kcat = []
        else:
            KCAT.append('N/A')
        k8 += 1
    if k9 < k1-1:
        if vmax != []:
            VMAX.append(vmax)
            vmax = []
        else:
            VMAX.append('N/A')
        k9 += 1
    if k10 < k1-1:
        if ph != []:
            PH.append(ph)
            ph = []
        else:
            PH.append('N/A')
        k10 += 1
    if k11 < k1-1:
        act.append('N/A')
        k11 += 1
    if k12 < k1-1:
        temp.append('N/A')
        k12 += 1

# double check row count

if k4 < k1:
    if reaction != []:
        rxn.append(reaction)
        reaction = []
    else:
        rxn.append('N/A')
    k4 += 1
if k5 < k1:
    #print(substrate)
    if substrate != []:
        alt.append(substrate)
        substrate = []
    else:
        alt.append('N/A')
    k5 += 1

if k6 < k1:
    if subs != []:
        sub.append(subs)
        subs = []
    else:
        sub.append('N/A')
    k6 += 1
    
if k7 < k1:
    if km != []:
        KM.append(km)
        km = []
    else:
        KM.append('N/A')
    k7 += 1
if k8 < k1:
    if kcat != []:
        KCAT.append(kcat)
        kcat = []
    else:
        KCAT.append('N/A')
    k8 += 1
if k9 < k1:
    if vmax != []:
        VMAX.append(vmax)
        vmax = []
    else:
        VMAX.append('N/A')
    k9 += 1
if k10 < k1:
    if ph != []:
        PH.append(ph)
        ph = []
    else:
        PH.append('N/A')
    k10 += 1
if k11 < k1:
    act.append('N/A')
    k11 += 1
if k12 < k1:
    temp.append('N/A')
    k12 += 1

print ('Number of compounds count is', k1, k2, k3, 'rxn', k4, 'alt-subs', k5, '^substrate', k6, k7, k8, k9, k10, k11, k12)
```

```python
df_enzrxns = pd.DataFrame({'UNIQUE-ID' : ID,
                           'COMMON-NAME' : CName,
                           'ENZYME' : enz,
                           'REACTION' : rxn,
                           'ALTERNATIVE-SUBSTRATES' : alt,
                           '^SUBSTRATE' : sub,
                           'KM' : KM,
                           'KCAT' : KCAT,
                           'VMAX' : VMAX,
                           'PH-OPT' : PH,
                           'SPECIFIC-ACTIVITY' : act,
                           'TEMPERATURE-OPT' : temp
                          })
```
